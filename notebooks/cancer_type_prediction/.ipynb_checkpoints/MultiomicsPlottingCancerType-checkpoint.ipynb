# ROC

```python
from sklearn.preprocessing import label_binarize
```

```python
dl_alltypes_df = pd.read_csv(
    "../../work_dirs/tcga_all_cancer_types//DeepMultiOmicNet/mutation_cnv_rna/scores_202201272003_DOIT_LRP_23_types.csv.gz"
)
dl_alltypes_all_pred_df = pd.read_csv(
    "../../work_dirs/tcga_all_cancer_types/DeepMultiOmicNet/mutation_cnv_rna/all_val_res_202201272003_DOIT_LRP_23_types.csv.gz"
)
dl_alltypes_df_max = dl_alltypes_df.groupby('run').max().reset_index().rename(
    columns={'top1_acc': 'acc'})[['run', 'acc', 'f1', 'roc_auc']]
```

```python
dl_alltypes_df_max_auc = pd.merge(dl_alltypes_df_max[['run', 'roc_auc']],
                              dl_alltypes_df[['run', 'roc_auc', 'epoch']],
                              on=['run', 'roc_auc'])
```

```python
dl_alltypes_conf_df = pd.merge(dl_alltypes_df_max_auc, dl_alltypes_all_pred_df, on=['run','epoch'])
```

```python
y_true = dl_alltypes_conf_df['y_true']
n_classes = 23
y_score = dl_alltypes_conf_df.iloc[:,-n_classes:]
class_names = dl_alltypes_conf_df.iloc[:,-n_classes:].columns
```

```python
alltypes_palette = dict(
    zip(class_names, [
        "#00825f", "#792ab2", "#ff9fd3", "#7b3a86", "#015dc9", "#aaec2a",
        "#be4500", "#7c6fff", "#e3bbc1", "#ff66fc", "#bf0096", "#ff8966",
        "#ffde5e", "#52c91a", "#66ffb6", "#655315", "#8b9900", "#00d8fe",
        "#008601", "#8aa8ff", "#a3184c", "#fff62d", "#c29700"
    ]))
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for run in dl_alltypes_conf_df['run'].unique():
    roc_auc[run] = dict()
    fpr[run] = dict()
    tpr[run] = dict()

    y_true_tmp = label_binarize(y_true[dl_alltypes_conf_df['run'] == run],
                                classes=class_names)
    y_score_tmp = y_score[dl_alltypes_conf_df['run'] == run].values
    for i in range(len(class_names)):
        fpr[run][class_names[i]], tpr[run][class_names[i]], _ = roc_curve(
            y_true_tmp[:, i], y_score_tmp[:, i])
        roc_auc[run][class_names[i]] = auc(fpr[run][class_names[i]],
                                           tpr[run][class_names[i]])
    fpr[run]["micro"], tpr[run]["micro"], _ = roc_curve(
        y_true_tmp.ravel(), y_score_tmp.ravel())
    roc_auc[run]["micro"] = auc(fpr[run]["micro"], tpr[run]["micro"])

    all_fpr = np.unique(
        np.concatenate([fpr[run][class_names[i]] for i in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[run][class_names[i]],
                              tpr[run][class_names[i]])
    mean_tpr /= n_classes

    fpr[run]["macro"] = all_fpr
    tpr[run]["macro"] = mean_tpr
    roc_auc[run]["macro"] = auc(fpr[run]["macro"], tpr[run]["macro"])
```

```python
sd_auc = dict()
for c in roc_auc['cv_0'].keys():
    tmp_auc = [roc_auc[run][c] for run in dl_alltypes_conf_df['run'].unique()]
    sd_auc[c] = np.std(tmp_auc)

fpr_flattened = dict()
tpr_flattened = dict()
auc_summary = dict()
for c in roc_auc['cv_0'].keys():
    fpr_flattened[c] = []
    tpr_flattened[c] = []
    auc_summary[c] = []
    for run in dl_alltypes_conf_df['run'].unique():
        fpr_flattened[c].extend(fpr[run][c])
        tpr_flattened[c].extend(tpr[run][c])
        auc_summary[c].append(roc_auc[run][c])
    fpr_flattened[c] = sorted(fpr_flattened[c])
    tpr_flattened[c] = sorted(tpr_flattened[c])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2.7, 2.7))
for c in class_names:
    lw = 0.5
    ax.plot(
        fpr_flattened[c],
        tpr_flattened[c],
        color=alltypes_palette[c],
        lw=lw,
        label=
        f"{c} (AUROC = {np.mean(auc_summary[c]):.2g}$\pm${sem(auc_summary[c]):.1g})",
    )
    ax.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
    ax.set_xlim([-0.01, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel("1 - Specificity")
    ax.set_ylabel("Sensitivity")
    ax.legend(bbox_to_anchor=(1.03, 1), fontsize=4.5, loc=2, borderaxespad=0.)
plt.title("All classes ROC curves")
# plt.tight_layout()
plt.savefig("../../supp_plots/23_roc.pdf", dpi=300)
```

## PR

```python
from sklearn.metrics import average_precision_score, precision_recall_curve
```

```python
precision = dict()
recall = dict()
ap = dict()
for run in dl_alltypes_conf_df['run'].unique():
    ap[run] = dict()
    precision[run] = dict()
    recall[run] = dict()

    y_true_tmp = label_binarize(y_true[dl_alltypes_conf_df['run'] == run],
                                classes=class_names)
    y_score_tmp = y_score[dl_alltypes_conf_df['run'] == run].values
    for i in range(len(class_names)):
        precision[run][class_names[i]], recall[run][
            class_names[i]], _ = precision_recall_curve(
                y_true_tmp[:, i], y_score_tmp[:, i])
        ap[run][class_names[i]] = average_precision_score(
            y_true_tmp[:, i], y_score_tmp[:, i])
    precision[run]["micro"], recall[run]["micro"], _ = precision_recall_curve(
        y_true_tmp.ravel(), y_score_tmp.ravel())
    ap[run]["micro"] = average_precision_score(y_true_tmp.ravel(),
                                               y_score_tmp.ravel())

    all_recall = np.unique(
        np.concatenate([recall[run][class_names[i]]
                        for i in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_precision = np.zeros_like(all_recall)
    for i in range(n_classes):
        mean_precision += np.interp(all_recall, sorted(recall[run][class_names[i]], reverse=False),
                                    sorted(precision[run][class_names[i]], reverse=True))
    mean_precision /= n_classes

    recall[run]["macro"] = all_recall
    precision[run]["macro"] = mean_precision

    all_ap = np.array([ap[run][c] for c in class_names])
    # Then interpolate all ROC curves at this points
    ap[run]["macro"] = all_ap.mean()
```

```python
precision_flattened = dict()
recall_flattened = dict()
ap_summary = dict()
for c in ap['cv_0'].keys():
    precision_flattened[c] = []
    recall_flattened[c] = []
    ap_summary[c] = []
    for run in dl_alltypes_conf_df['run'].unique():
        precision_flattened[c].extend(precision[run][c])
        recall_flattened[c].extend(recall[run][c])
        ap_summary[c].append(ap[run][c])
    precision_flattened[c] = sorted(precision_flattened[c])
    recall_flattened[c] = sorted(recall_flattened[c])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2.7, 2.7))
for c in class_names:
    lw = 0.5
    ax.plot(
        sorted(recall_flattened[c], reverse=True),
        precision_flattened[c],
        color=alltypes_palette[c],
        lw=lw,
        label=
        f"{c} (AUPRC = {np.mean(ap_summary[c]):.2g}$\pm${sem(ap_summary[c]):.1g})",
    )
    ax.set_xlim([0, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
#     ax.legend(loc="lower right", fontsize=5)
plt.legend(bbox_to_anchor=(1.03, 1), fontsize=4.5, loc=2, borderaxespad=0.)
plt.title("All classes Pecision-Recall curve")
plt.savefig("../../supp_plots/23_pr.pdf")
```

## ROC

```python
from sklearn.preprocessing import label_binarize
```

```python
dl_brca_df = pd.read_csv(
    "../../work_dirs/tcga_brca_subtypes/DeepMultiOmicNet/mutation_cnv_rna/scores_202201272059_DOIT_LRP.csv.gz"
)
dl_brca_all_pred_df = pd.read_csv(
    "../../work_dirs/tcga_brca_subtypes/DeepMultiOmicNet/mutation_cnv_rna/all_val_res_202201272059_DOIT_LRP.csv.gz"
)
dl_brca_df_max = dl_brca_df.groupby('run').max().reset_index().rename(
    columns={'top1_acc': 'acc'})[['run', 'acc', 'f1', 'roc_auc']]
dl_brca_all_pred_df['y_pred'] = dl_brca_all_pred_df['y_pred'].map(
    lambda x: x.replace('BRCA.', ''))
dl_brca_all_pred_df['y_true'] = dl_brca_all_pred_df['y_true'].map(
    lambda x: x.replace('BRCA.', ''))
dl_brca_all_pred_df.columns = [
    x.replace('BRCA.', '') for x in dl_brca_all_pred_df.columns
]
```

```python
dl_brca_df_max_auc = pd.merge(dl_brca_df_max[['run', 'roc_auc']],
                              dl_brca_df[['run', 'roc_auc', 'epoch']],
                              on=['run', 'roc_auc'])
```

```python
dl_brca_conf_df = pd.merge(dl_brca_df_max_auc, dl_brca_all_pred_df, on=['run','epoch'])
```

```python
y_true = dl_brca_conf_df['y_true']
n_classes = 5
y_score = dl_brca_conf_df.iloc[:,-n_classes:]
class_names = dl_brca_conf_df.iloc[:,-n_classes:].columns
```

```python
brca_palette = dict(
    zip(class_names, ["#8d8ae8", "#c6c04d", "#e46ab8", "#6ecf79", "#e2784c"]))
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for run in dl_brca_conf_df['run'].unique():
    roc_auc[run] = dict()
    fpr[run] = dict()
    tpr[run] = dict()

    y_true_tmp = label_binarize(y_true[dl_brca_conf_df['run'] == run],
                                classes=class_names)
    y_score_tmp = y_score[dl_brca_conf_df['run'] == run].values
    for i in range(len(class_names)):
        fpr[run][class_names[i]], tpr[run][class_names[i]], _ = roc_curve(
            y_true_tmp[:, i], y_score_tmp[:, i])
        roc_auc[run][class_names[i]] = auc(fpr[run][class_names[i]],
                                           tpr[run][class_names[i]])
    fpr[run]["micro"], tpr[run]["micro"], _ = roc_curve(
        y_true_tmp.ravel(), y_score_tmp.ravel())
    roc_auc[run]["micro"] = auc(fpr[run]["micro"], tpr[run]["micro"])

    all_fpr = np.unique(
        np.concatenate([fpr[run][class_names[i]] for i in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[run][class_names[i]],
                              tpr[run][class_names[i]])
    mean_tpr /= n_classes

    fpr[run]["macro"] = all_fpr
    tpr[run]["macro"] = mean_tpr
    roc_auc[run]["macro"] = auc(fpr[run]["macro"], tpr[run]["macro"])
```

```python
sd_auc = dict()
for c in roc_auc['cv_0'].keys():
    tmp_auc = [roc_auc[run][c] for run in dl_brca_conf_df['run'].unique()]
    sd_auc[c] = np.std(tmp_auc)

fpr_flattened = dict()
tpr_flattened = dict()
auc_summary = dict()
for c in roc_auc['cv_0'].keys():
    fpr_flattened[c] = []
    tpr_flattened[c] = []
    auc_summary[c] = []
    for run in dl_brca_conf_df['run'].unique():
        fpr_flattened[c].extend(fpr[run][c])
        tpr_flattened[c].extend(tpr[run][c])
        auc_summary[c].append(roc_auc[run][c])
    fpr_flattened[c] = sorted(fpr_flattened[c])
    tpr_flattened[c] = sorted(tpr_flattened[c])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2.7, 2.7))
for c in class_names:
    lw = 0.8
    ax.plot(
        fpr_flattened[c],
        tpr_flattened[c],
        color=brca_palette[c],
        lw=lw,
        label=
        f"{c} (AUROC = {np.mean(auc_summary[c]):.2g}$\pm${sem(auc_summary[c]):.1g})",
    )
    ax.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
    ax.set_xlim([-0.01, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel("1 - Specificity")
    ax.set_ylabel("Sensitivity")
    ax.legend(loc="lower right", fontsize=6)

plt.title("All classes ROC curves")
plt.savefig("../../supp_plots/brca_roc.pdf")
```

## PR

```python
from sklearn.metrics import average_precision_score, precision_recall_curve
```

```python
precision = dict()
recall = dict()
ap = dict()
for run in dl_brca_conf_df['run'].unique():
    ap[run] = dict()
    precision[run] = dict()
    recall[run] = dict()

    y_true_tmp = label_binarize(y_true[dl_brca_conf_df['run'] == run],
                                classes=class_names)
    y_score_tmp = y_score[dl_brca_conf_df['run'] == run].values
    for i in range(len(class_names)):
        precision[run][class_names[i]], recall[run][
            class_names[i]], _ = precision_recall_curve(
                y_true_tmp[:, i], y_score_tmp[:, i])
        ap[run][class_names[i]] = average_precision_score(
            y_true_tmp[:, i], y_score_tmp[:, i])
    precision[run]["micro"], recall[run]["micro"], _ = precision_recall_curve(
        y_true_tmp.ravel(), y_score_tmp.ravel())
    ap[run]["micro"] = average_precision_score(y_true_tmp.ravel(),
                                               y_score_tmp.ravel())

    all_recall = np.unique(
        np.concatenate([recall[run][class_names[i]]
                        for i in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_precision = np.zeros_like(all_recall)
    for i in range(n_classes):
        mean_precision += np.interp(all_recall, sorted(recall[run][class_names[i]], reverse=False),
                                    sorted(precision[run][class_names[i]], reverse=True))
    mean_precision /= n_classes

    recall[run]["macro"] = all_recall
    precision[run]["macro"] = mean_precision

    all_ap = np.array([ap[run][c] for c in class_names])
    # Then interpolate all ROC curves at this points
    ap[run]["macro"] = all_ap.mean()
```

```python
precision_flattened = dict()
recall_flattened = dict()
ap_summary = dict()
for c in ap['cv_0'].keys():
    precision_flattened[c] = []
    recall_flattened[c] = []
    ap_summary[c] = []
    for run in dl_brca_conf_df['run'].unique():
        precision_flattened[c].extend(precision[run][c])
        recall_flattened[c].extend(recall[run][c])
        ap_summary[c].append(ap[run][c])
    precision_flattened[c] = sorted(precision_flattened[c])
    recall_flattened[c] = sorted(recall_flattened[c])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2.7, 2.7))
for c in class_names:
    lw = 0.5
    ax.plot(
        sorted(recall_flattened[c], reverse=True),
        precision_flattened[c],
        color=brca_palette[c],
        lw=lw,
        label=
        f"{c} (AUPRC = {np.mean(ap_summary[c]):.2g}$\pm${sem(ap_summary[c]):.1g})",
    )
    ax.set_xlim([0, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
    ax.legend(loc="lower left", fontsize=6)
# plt.legend(bbox_to_anchor=(1.02, 1), fontsize=6, loc=2, borderaxespad=0.)
# plt.title("All classes Pecision-Recall curve")
plt.savefig("../../supp_plots/brca_pr.pdf")
```

```python
y_true = dl_brca_val_conf_df['y_true']
n_classes = 5
y_score = dl_brca_val_conf_df.iloc[:,-n_classes:]
class_names = dl_brca_val_conf_df.iloc[:,-n_classes:].columns
```

```python
brca_palette = dict(
    zip(class_names, ["#8d8ae8", "#c6c04d", "#e46ab8", "#6ecf79", "#e2784c"]))
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()
for run in dl_brca_val_conf_df['run'].unique():
    roc_auc[run] = dict()
    fpr[run] = dict()
    tpr[run] = dict()

    y_true_tmp = label_binarize(y_true[dl_brca_val_conf_df['run'] == run],
                                classes=class_names)
    y_score_tmp = y_score[dl_brca_val_conf_df['run'] == run].values
    for i in range(len(class_names)):
        fpr[run][class_names[i]], tpr[run][class_names[i]], _ = roc_curve(
            y_true_tmp[:, i], y_score_tmp[:, i])
        roc_auc[run][class_names[i]] = auc(fpr[run][class_names[i]],
                                           tpr[run][class_names[i]])
    fpr[run]["micro"], tpr[run]["micro"], _ = roc_curve(
        y_true_tmp.ravel(), y_score_tmp.ravel())
    roc_auc[run]["micro"] = auc(fpr[run]["micro"], tpr[run]["micro"])

    all_fpr = np.unique(
        np.concatenate([fpr[run][class_names[i]] for i in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += np.interp(all_fpr, fpr[run][class_names[i]],
                              tpr[run][class_names[i]])
    mean_tpr /= n_classes

    fpr[run]["macro"] = all_fpr
    tpr[run]["macro"] = mean_tpr
    roc_auc[run]["macro"] = auc(fpr[run]["macro"], tpr[run]["macro"])
```

```python
sd_auc = dict()
for c in roc_auc['cv_0'].keys():
    tmp_auc = [roc_auc[run][c] for run in dl_brca_val_conf_df['run'].unique()]
    sd_auc[c] = np.std(tmp_auc)

fpr_flattened = dict()
tpr_flattened = dict()
auc_summary = dict()
for c in roc_auc['cv_0'].keys():
    fpr_flattened[c] = []
    tpr_flattened[c] = []
    auc_summary[c] = []
    for run in dl_brca_val_conf_df['run'].unique():
        fpr_flattened[c].extend(fpr[run][c])
        tpr_flattened[c].extend(tpr[run][c])
        auc_summary[c].append(roc_auc[run][c])
    fpr_flattened[c] = sorted(fpr_flattened[c])
    tpr_flattened[c] = sorted(tpr_flattened[c])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2, 2))
for c in class_names:
    lw = 0.8
    ax.plot(
        fpr_flattened[c],
        tpr_flattened[c],
        color=brca_palette[c],
        lw=lw,
        label=f"{c} (AUROC = {np.mean(auc_summary[c]):.2g})",
    )
    ax.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
    ax.set_xlim([-0.01, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel("1 - Specificity")
    ax.set_ylabel("Sensitivity")
    ax.legend(loc="lower right", fontsize=6)

plt.title("All classes ROC curves", fontdict={'size': 7})
plt.savefig("../../plots/brca_validation_roc.pdf", dpi=300)
```

## PR

```python
from sklearn.metrics import average_precision_score, precision_recall_curve
```

```python
precision = dict()
recall = dict()
ap = dict()
for run in dl_brca_val_conf_df['run'].unique():
    ap[run] = dict()
    precision[run] = dict()
    recall[run] = dict()

    y_true_tmp = label_binarize(y_true[dl_brca_val_conf_df['run'] == run],
                                classes=class_names)
    y_score_tmp = y_score[dl_brca_val_conf_df['run'] == run].values
    for i in range(len(class_names)):
        precision[run][class_names[i]], recall[run][
            class_names[i]], _ = precision_recall_curve(
                y_true_tmp[:, i], y_score_tmp[:, i])
        ap[run][class_names[i]] = average_precision_score(
            y_true_tmp[:, i], y_score_tmp[:, i])
    precision[run]["micro"], recall[run]["micro"], _ = precision_recall_curve(
        y_true_tmp.ravel(), y_score_tmp.ravel())
    ap[run]["micro"] = average_precision_score(y_true_tmp.ravel(),
                                               y_score_tmp.ravel())

    all_recall = np.unique(
        np.concatenate([recall[run][class_names[i]]
                        for i in range(n_classes)]))
    # Then interpolate all ROC curves at this points
    mean_precision = np.zeros_like(all_recall)
    for i in range(n_classes):
        mean_precision += np.interp(all_recall, sorted(recall[run][class_names[i]], reverse=False),
                                    sorted(precision[run][class_names[i]], reverse=True))
    mean_precision /= n_classes

    recall[run]["macro"] = all_recall
    precision[run]["macro"] = mean_precision

    all_ap = np.array([ap[run][c] for c in class_names])
    # Then interpolate all ROC curves at this points
    ap[run]["macro"] = all_ap.mean()
```

```python
precision_flattened = dict()
recall_flattened = dict()
ap_summary = dict()
for c in class_names:
    precision_flattened[c] = []
    recall_flattened[c] = []
    ap_summary[c] = []
    for run in dl_brca_val_conf_df['run'].unique():
        precision_flattened[c].extend(precision[run][c])
        recall_flattened[c].extend(recall[run][c])
        ap_summary[c].append(ap[run][c])
    precision_flattened[c] = sorted(precision_flattened[c])
    recall_flattened[c] = sorted(recall_flattened[c])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(2, 2))
for c in class_names:
    lw = 0.8
    ax.plot(
        sorted(recall_flattened[c], reverse=True),
        precision_flattened[c],
        color=brca_palette[c],
        lw=lw,
        label=f"{c} (AUPRC = {np.mean(ap_summary[c]):.2g})",
    )
    ax.set_xlim([0, 1.0])
    ax.set_ylim([0.0, 1.02])
    ax.set_xlabel("Recall")
    ax.set_ylabel("Precision")
    ax.legend(loc="lower left", fontsize=6)
# plt.legend(bbox_to_anchor=(1.02, 1), fontsize=6, loc=2, borderaxespad=0.)
# plt.title("All classes Pecision-Recall curve")
# plt.show()
plt.savefig("../../plots/brca_validation_pr.pdf", dpi=300)
```
