```python
def corrfunc(x,y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    # Unicode for lowercase rho (œÅ)
#     rho = '\u03C1'
    rho = 'R'
    ax.annotate(f'{rho} = {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes, fontsize=6)
    ax.plot((0, 0.6), (0, 0.6), color='red', zorder=10, linewidth=0.25)
```

```python
def medianfunc(x, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    median = np.median(x)
    ax = ax or plt.gca()
    ax.annotate(f'median: {median:.4f}', xy=(.5, .9), xycoords=ax.transAxes, fontsize=6)
```

# Sanger GDSC

```python
en_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/ec_en/mutation_cnv_rna/scores_202201041912_ec_en_allgenes_drug_mutation_cnv_rna.csv"
).groupby(['drug_id']).mean()
en_mutation_cnv_rna['model'] = 'elastic net'

rf_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/ec_rf/mutation_cnv_rna/scores_202201080129_ec_rf_allgenes_drug_mutation_cnv_rna.csv"
).groupby(['drug_id']).mean()
rf_mutation_cnv_rna['model'] = 'random forest'

mixomics_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/mixOmics/sanger_mutation_cnv_rna_gdsc_50_comp_allgenes.csv"
).groupby(['drug_id']).mean()
mixomics_mutation_cnv_rna[
    'model'] = 'mixOmics'

pca_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/pca_rf/mutation_cnv_rna/scores_202201052331_pca_rf_allgenes_drug_mutation_cnv_rna.csv"
).groupby(['drug_id']).mean()
pca_mutation_cnv_rna[
    'model'] = 'PCA'

moCluster_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/moCluster_rf/mutation_cnv_rna/scores_202201052331_moCluster_rf_allgenes_drug_mutation_cnv_rna.csv"
).groupby(['drug_id']).median()
moCluster_mutation_cnv_rna['model'] = 'moCluster'

dl_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/DeepMultiOmicNet/mutation_cnv_rna/scores_202107081045_DOIT_LRP.csv.gz"
)

dl_list = [dl_mutation_cnv_rna.groupby(['drug_id',
                     'run']).max()['corr'].reset_index().groupby('drug_id').mean(),
           dl_mutation_cnv_rna.groupby(['drug_id',
                     'run']).max()['r2'].reset_index().groupby('drug_id').mean(),
           dl_mutation_cnv_rna.groupby(['drug_id',
                     'run']).min()['mae'].reset_index().groupby('drug_id').mean(),
           dl_mutation_cnv_rna.groupby(['drug_id',
                     'run']).min()['rmse'].reset_index().groupby('drug_id').mean()]
dl_merged = pd.concat(dl_list, axis=1)
dl_merged['model'] = 'DeePathNet'
```

```python
mixomics_mutation_cnv_rna.index = rf_mutation_cnv_rna.index
```

```python
cols = ['drug_id', 'corr', 'r2', 'mae', 'rmse', 'model']
commbined_df = pd.concat([
    dl_merged.reset_index()[cols],
    rf_mutation_cnv_rna.reset_index()[cols],
    en_mutation_cnv_rna.reset_index()[cols],
    pca_mutation_cnv_rna.reset_index()[cols],
    moCluster_mutation_cnv_rna.reset_index()[cols],
    mixomics_mutation_cnv_rna.reset_index()[cols]
])
```

```python
commbined_df_r2 = pd.pivot(commbined_df[['drug_id', 'model', 'r2']], index='drug_id', columns='model')

commbined_df_r2.columns = [x[1] for x in commbined_df_r2.columns]
```

```python
fig, ax = plt.subplots(1, 5, figsize=(8.5, 1.5), sharey=True, sharex=True)
for i in range(1, 6):
    sns.scatterplot(x=commbined_df_r2.columns[i],
                    y='DeePathNet',
                    data=commbined_df_r2,
                    ax=ax[i - 1],
                    s=4)
    r, p = pearsonr(commbined_df_r2.iloc[:, i], commbined_df_r2.iloc[:, 0])
    rho = 'R'
    ax[i - 1].annotate(f'{rho} = {r:.2f}, p={p:.2e}',
                       xy=(.1, .9),
                       xycoords=ax[i - 1].transAxes,
                       fontsize=6)
    ax[i - 1].plot((0, 0.6), (0, 0.6), color='red', zorder=10, linewidth=0.25)
plt.savefig("../../supp_plots/supp2_sanger_gdsc_consistency.pdf", dpi=500)
```

```python
g = sns.pairplot(commbined_df_r2, corner=False, plot_kws={'s': 4})
g.map_diag(medianfunc)
g.map_lower(corrfunc)
g.map_upper(corrfunc)
g.fig.set_size_inches(7, 7)

plt.savefig("../../supp_plots/supp2_sanger_gdsc_consistency.pdf", dpi=500)
```

## sanger + GDSC

```python
dl_sanger_gdsc_cancer = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/DeepMultiOmicNet/mutation_cnv_rna/scores_202107081045_DOIT_LRP.csv.gz"
)

dl_list = [
    dl_sanger_gdsc_cancer.groupby(['drug_id',
                                   'run']).max()['corr'].reset_index(),
    dl_sanger_gdsc_cancer.groupby(
        ['drug_id', 'run']).max()['r2'].reset_index().drop(['drug_id', 'run'],
                                                           axis=1),
    dl_sanger_gdsc_cancer.groupby(
        ['drug_id', 'run']).min()['mae'].reset_index().drop(['drug_id', 'run'],
                                                            axis=1),
    dl_sanger_gdsc_cancer.groupby(['drug_id',
                                   'run']).min()['rmse'].reset_index().drop(
                                       ['drug_id', 'run'], axis=1)
]
dl_merged_cancer = pd.concat(dl_list, axis=1)
dl_merged_cancer.columns = [
    f"{x}_cancergenes" for x in dl_merged_cancer.columns
]

dl_sanger_gdsc_all = pd.read_csv(
    f"../../work_dirs/sanger_gdsc_intersection_noprot/DeepMultiOmicNet/mutation_cnv_rna/scores_202201061531_all_genes.csv.gz"
)

dl_list = [
    dl_sanger_gdsc_all.groupby(['drug_id', 'run']).max()['corr'].reset_index(),
    dl_sanger_gdsc_all.groupby(
        ['drug_id', 'run']).max()['r2'].reset_index().drop(['drug_id', 'run'],
                                                           axis=1),
    dl_sanger_gdsc_all.groupby(
        ['drug_id', 'run']).min()['mae'].reset_index().drop(['drug_id', 'run'],
                                                            axis=1),
    dl_sanger_gdsc_all.groupby(['drug_id',
                                'run']).min()['rmse'].reset_index().drop(
                                    ['drug_id', 'run'], axis=1)
]
dl_merged_all = pd.concat(dl_list, axis=1)
dl_merged_all.columns = [
    f"{x}_allgenes" for x in dl_merged_all.columns
]
```

```python
dl_merged_sanger_gdsc = pd.concat([dl_merged_cancer, dl_merged_all],axis=1)
```

```python
data_range = [-0.09, 0.7]
plt.figure(figsize=(2, 2))
ax = sns.scatterplot(x='r2_cancergenes',
                y='r2_allgenes',
                data=dl_merged_sanger_gdsc,
                s=2)
plt.plot(data_range, data_range, c='red', lw=0.5)
plt.xlim(data_range)
plt.ylim(data_range)
r, _ = pearsonr(dl_merged_sanger_gdsc['r2_cancergenes'], dl_merged_sanger_gdsc['r2_allgenes'])
rho = 'R'
ax.annotate(f'{rho} = {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes, fontsize=6)
plt.xlabel("cancer-genes R$^{2}$")
plt.ylabel("all-genes R$^{2}$")
plt.tight_layout()
plt.savefig("../../supp_plots/cancer_vs_all_r2_sanger.png", dpi=300)
```

```python
data_range = [0.2, 3.1]
plt.figure(figsize=(2, 2))
ax = sns.scatterplot(x='mae_cancergenes',
                     y='mae_allgenes',
                     data=dl_merged_sanger_gdsc,
                     s=2)
plt.plot(data_range, data_range, c='red', lw=0.5)
plt.xlim(data_range)
plt.ylim(data_range)
r, _ = pearsonr(dl_merged_sanger_gdsc['mae_cancergenes'],
                dl_merged_sanger_gdsc['mae_allgenes'])
rho = 'R'
ax.annotate(f'{rho} = {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes, fontsize=6)
plt.xlabel("cancer-genes MAE")
plt.ylabel("all-genes MAE")
plt.tight_layout()
plt.savefig("../../supp_plots/cancer_vs_all_mae_sanger.png", dpi=300)
```
