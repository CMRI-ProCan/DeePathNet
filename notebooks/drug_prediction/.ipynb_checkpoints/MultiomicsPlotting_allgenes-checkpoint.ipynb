```python
pathway_palette = dict(
    zip(drug_summary_df['target_pathway'].unique(), [
        "#ffc9ff", "#c0cf83", "#99dea4", "#cfffe2", "#ffc9f0", "#30d1d6",
        "#68e1c8", "#ffcbac", "#87bdff", "#a4f3ff", "#aaa9c5", "#ffd2cf",
        "#77b4ca", "#ee9dd0", "#ffb790", "#cd9eb8", "#dcffbf", "#de9f6c",
        "#abceff", "#efba7d", "#75bb8b", "#c9a48a", "#ccbe73"
    ]))
```

```python
drug_summary_name_pathway = drug_summary_df.set_index(
    'drug_name').to_dict()['target_pathway']
```

```python
dl_mutation_cnv_rna = pd.read_csv(
    f"../../work_dirs/sanger_train_ccle_test_gdsc/DeePathNet/mutation_cnv_rna_prot/scores_202205281134_DOIT_LRP.csv.gz"
)

dl_list = [
    dl_mutation_cnv_rna.groupby(['drug_id',
                                 'run']).max()['corr'].reset_index(),
    dl_mutation_cnv_rna.groupby(
        ['drug_id', 'run']).max()['r2'].reset_index().drop(['drug_id', 'run'],
                                                           axis=1),
    dl_mutation_cnv_rna.groupby(
        ['drug_id', 'run']).min()['mae'].reset_index().drop(['drug_id', 'run'],
                                                            axis=1),
    dl_mutation_cnv_rna.groupby(['drug_id',
                                 'run']).min()['rmse'].reset_index().drop(
                                     ['drug_id', 'run'], axis=1)
]
dl_merged_cancer = pd.concat(dl_list, axis=1)
```

```python
dl_merged_cancer['Target Pathway'] = dl_merged_cancer['drug_id'].map(
    lambda x: drug_summary_name_pathway[x.split(";")[1]])
```

```python
dl_merged_cancer['Drug Name'] = dl_merged_cancer['drug_id'].map(
    lambda x: x.split(";")[1])
```

```python
r2_top = dl_merged_cancer.sort_values(
    by='r2', ascending=False).head(20).reset_index(drop=True)
mae_top = dl_merged_cancer.sort_values(
    by='mae', ascending=True).head(20).reset_index(drop=True)
corr_top = dl_merged_cancer.sort_values(
    by='corr', ascending=False).head(20).reset_index(drop=True)
```

```python
plt.figure(figsize=(4,1.5))
ax1 = sns.barplot(x='Drug Name',
                  y='corr',
                  dodge=False,
                  color='skyblue',
                  hue='Target Pathway',
                  palette=pathway_palette,
                  data=corr_top)
plt.legend("").remove()
plt.setp(ax1.get_xticklabels(),
         rotation=45,
         ha="right",
         rotation_mode="anchor",
         fontsize=5)
print()
plt.savefig("../../supp_plots/top_drug_corr.pdf")
```
